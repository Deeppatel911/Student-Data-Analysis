columns like "INTERNET", "TV", "COMPUTER", "WASHING_MCH", "MIC_OVEN", "CAR", "DVD", "FRESH", "PHONE" and "MOBILE" contains values
like "YES" and "NO" we can convert into "0" and "1".

```python
things = ["INTERNET", "TV", "COMPUTER", "WASHING_MCH", "MIC_OVEN", "CAR", "DVD", "FRESH", "PHONE","MOBILE"]
for label in things:
    df[label] = pd.DataFrame(pd.get_dummies(df[label]).values,columns=["NO","YES"])["YES"]
```

```python
df = df.drop(["COD_S11","Cod_SPro"], axis = 1)
df.head()
```

Here the "Unnamed: 9" is meaningless in the context of the database and does not contain even one value therefor it is in benefit to remove the column.

```python
df = df.drop(["Unnamed: 9"],axis = 1)
```

```python
df.head()
```

Atfer that, we can say that the data does not contain the null values or empty string but we need to filter the outlier in the database.

```python
columns = list(df.columns)
for thing in things:
    columns.remove(thing)
```

Here, the null values are represent by 0 so we can replace those by the most frequest item in the corrosponding column.
by ruuning the following loop through the data...

```python
for i in columns:
    df[i] = df[i].replace(0,df[i].mode()[0])
```

We also found the word "Ninguno" which is a spanish word which means None.we will delete those record in the databse.

```python
for i in df.columns:
    df = df[df[i] != "Ninguno"]
```

```python
df.PEOPLE_HOUSE.value_counts()
```

Here we can see the typo in PEOPLE_HOUSE column. we can replace by the right word.
Like spanish word "Nueve" means Nine in English.
Likewise we can replace "Once" to "One".

```python
df.PEOPLE_HOUSE = df.PEOPLE_HOUSE.replace("Nueve","Nine")
df.PEOPLE_HOUSE = df.PEOPLE_HOUSE.replace("Once","One")
```

```python
total_S11 = []  #total of SABER S11
total_PRO = []  #total of SABER PRO

for i, j, k, l, m in zip(df['MAT_S11'], df['CR_S11'], df['CC_S11'], df['BIO_S11'], df['ENG_S11']):
    total_S11.append(i+j+k+l+m)
    
for i, j, k, l, m in zip(df['QR_PRO'], df['CR_PRO'], df['CC_PRO'], df['ENG_PRO'], df['WC_PRO']):
    total_PRO.append(i+j+k+l+m)
    
df['TOTAL_S11'] = total_S11
df['TOTAL_PRO'] = total_PRO
```
